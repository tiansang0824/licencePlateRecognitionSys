# PlateLocator 类测试

该测试旨在检查PlateLocator类能否按照需要的方式进行工作。



```python
# 导包
import cv2 as cv

from source import PlateLocator
from source import ToolKit as toolKit

```

```python
# 定义图片路径，以及创建PlateLocator实例
original_image_path = '../resources/test_plate.png'  # 测试图片
plate_locator = PlateLocator.PlateLocator(cv.imread(original_image_path))  # 定义一个PlateLocator类
```

## 类函数逐个测试

下面依次调用类中的函数，实现对每一个处理步骤的功能性测试。


```python
# 输出类内保存的原图
toolKit.show_color('存入类中的原图', plate_locator.original_image)
```

下面开始调用函数


```python
# 原始图片
toolKit.show_color('原始图片',plate_locator.original_image)

# 高斯去噪
plate_locator.gauss_denoise()
toolKit.show_color('高斯去噪', plate_locator.gauss_image)
# 灰度处理
plate_locator.grayscale_process()
toolKit.show_gray('灰度处理', plate_locator.gray_image)
# 边缘检测
plate_locator.edge_detect()
toolKit.show_gray('边缘检测结果', plate_locator.abs_x)

# 阈值处理
plate_locator.adaptive_threshold()
toolKit.show_gray('阈值处理结果', plate_locator.abs_x)

# 闭运算、去除白点
plate_locator.closed_operation()
toolKit.show_gray('闭运算处理结果', plate_locator.closed_operated_image)

# 中值滤波
plate_locator.median_filter()
toolKit.show_gray('中值滤波', plate_locator.median_image)

# 轮廓检测
plate_locator.detect_contours(plate_locator.median_image)
toolKit.show_gray('（全部）轮廓检测结果', plate_locator.image_with_contours)

# 筛选车牌位置
plate_locator.find_plate_contour(plate_locator.contours)
toolKit.show_color('车牌轮廓位置', plate_locator.image_with_contour)

# 拟合直线
plate_locator.fit_straight_line(plate_locator.contour)
toolKit.show_color('拟合直线结果', plate_locator.image_with_line)

# 旋转图片
plate_locator.rotate_image(plate_locator.line_info)
toolKit.show_color('旋转图片结果', plate_locator.rotated_image)
```

利用预处理函数解决车牌倾斜问题。


```python
# 重新载入图片
original_image = cv.imread(original_image_path)
pl2 = PlateLocator.PlateLocator(original_image)

plate_locator.pre_process()

toolKit.show_color('预处理后的旋转图片',pl2.gauss_image)
```
